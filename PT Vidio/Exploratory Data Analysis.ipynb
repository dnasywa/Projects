# Gain Inisght

## Popular Platform by SQL

```python
# urutan platform paling populer,
# beserta jumlah kunjungan dan rata-rata pemutaran konten

cursor.execute(
    '''
    SELECT platform, COUNT(*) AS total_visit
    FROM stream_info
    GROUP BY platform
    ORDER BY total_visit DESC;
    '''
).fetchall()
```

## Advanced Discover by Python

Melanjutkan penemuan fakta bahwa Web Mobile adalah platform dengan total kunjungan terbanyak, pada bagian ini akan digali lebih dalam faktor yang potensial dalam mewujudkan hal tersebut

```python
data['platform'].value_counts().plot(kind='bar', ylabel='Total Visit', xlabel='Platform', color='#ED203F')
```

Didapatkan bahwa popularitas Web-Mobile sangatlah signifikan dibandingkan platform lainnya

```python
import matplotlib.pyplot as plt

mobile_web_data = data[data['platform'].isin(['web-mobile'])]
corr = mobile_web_data.corr()
plt.figure(figsize=(12, 10))

sns.heatmap(
    corr[(corr >= 0.5) | (corr <= -0.4)],
    cmap="coolwarm",
    vmax=1.0,
    vmin=-1.0,
    linewidths=0.1,
    annot=True,
    annot_kws={"size": 6},
    square=True
);
```

Didapatkan bahwa terhadap korelasi unik antara total byte serta durasi konten yang dikonsumsi pengguna platform Web Mobile

### Correlation: Play Duration

```python
average_streams = data.groupby('platform').play_duration.mean().sort_values(ascending=False)
average_streams.plot(kind='bar', xlabel='Platform', ylabel='Play Duration (minutes mean)', color='#ED203F')
```

Didapatkan bahwa rata-rata durasi konten yang dikonsumsi oleh pengguna platform Web Mobile cukup rendah, yaitu di bawah 200 menit atau kurang dari 4 jam untuk setiap kunjungannya

```python
average_content_duration = data.groupby('content_type').play_duration.mean().sort_values(ascending=True)
average_content_duration
```

Berdasarkan rata-rata waktu yang dihabiskan User untuk menonton konten di Vidio, tipe konten Video on Demand yang memakan 244 menit cukup logis untuk menjadi tontonan populer para pengguna platform Web Mobile

### Correlation: Total Bytes

```python
average_bytes = data.groupby('platform').total_bytes.mean().sort_values(ascending=False)
average_bytes.plot(kind='bar', xlabel='Platform', ylabel='Average Bytes', color='#ED203F')
```

Jumlah penggunaan Bytes oleh para user Web Mobile terbilang sangat rendah dibandingkan pada platform lainnya. Tentunya hal ini secara positif berkorelasi dengan durasi menonton yang juga rendah.
